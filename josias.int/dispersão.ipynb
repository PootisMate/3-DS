```python
import matplotlib.pyplot as plt
import numpy as np

investimento = np.random.randint(1000, 10000, 20)
vendas = investimento * 0.8 + np.random.randint(-1000, 1000, 20)

plt.figure(figsize=(8, 5))
plt.scatter(investimento, vendas, color='purple')

plt.xlabel("Investimento em Marketing (R$)")
plt.ylabel("Vendas (R$)")
plt.title("Relação entre Investimento em Vendas")
plt.show()
```
