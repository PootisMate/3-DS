```python
import matplotlib.pyplot as plt
import numpy as np

salarios = np.random.normal(loc=5000, scale=1500, size=200)

plt.figure(figsize=(8,5))
plt.hist(salarios, bins=15, color='blue', edgecolor='black', alpha=0.7)

plt.xlabel("Salário (R$)")
plt.ylabel("Frequência")
plt.title("Distribuição Salarial dos Funcionários")
plt.show()
```
